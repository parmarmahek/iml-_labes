```python
##mahek parmar
##216270307010
##Write a Pandas program to create a line plot of the opening, closing stock
##prices of given company between two specific dates.
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/HDFCBANK.csv')

df['Date'] = pd.to_datetime(df['Date'])

start_date = '2000-01-03'
end_date = '2000-03-13'

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Opening Price', color='blue')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Closing Price', color='red')

plt.title('Stock Prices between {} and {}'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```
