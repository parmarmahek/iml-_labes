```python
#216270307010
#parmar mahek
#to convert the first column of a DataFrame as a Series
import pandas as pd  
 
df = pd.DataFrame() 
print(df)
# list of strings 
lst = ['Geeks', 'For', 'Geeks', 'is', 'portal', 'for', 'Geeks'] 
 
df = pd.DataFrame(lst) 
print("Dataframe")
print(df)

```

```python
##216270307010
#parmar mahek
#to convert the first column of a DataFrame as a Series
import pandas as pd 
 
df = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,6]}) 
series = df.iloc[:, 0]
print(series)
```

```python
##216270307010
#parmar mahek
#to create the mean and standard deviation of the data of a given Series
import pandas as pd

s = pd.Series(data = [9, 9, 8, 5, 7, 8, 1, 2, 3, 4, 5, 6, 7, 8, 9, 5, 3])

print(s.mean())
print(s.std())

```
