```python
##216270307010
##mahek parmar 
#addition,substract,multyply,divide,argument element-wise
import numpy as np
x = np.arange(4)
print("x =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
    

```

```python
#216270307010
#mahek parmar 
#to  mean across dimension,in a 2d numpy array
import numpy as np
x=np.array([[6,8],[3,9]])
print("x:\n",x)
print("mean of x:",np.mean(x))
print("mean of a along axis=0:",np.mean(x,axis=0))
print("mean of a along axis=1:",np.mean(x,axis=1))


```
